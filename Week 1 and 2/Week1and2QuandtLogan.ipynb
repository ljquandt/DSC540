```python
#generate random number list
random_number_list = [random.randint(0, 100) for x in range (0,100)]
print(random_number_list)
```

```python
#create list of numbers divisible by 3
divisible_by_3_list = [a for a in random_number_list if a % 3 == 0]
divisible_by_3_list
```

```python
# find difference between lists
length_random = len(random_number_list)
length_div3 = len(divisible_by_3_list)
difference = length_random - length_div3
print(difference)
```

```python
#loop to calculate difference multiple times and add to list
number_of_tests = 3
difference_list = []
for i in range(0, number_of_tests):
    random_number_list2 = [random.randint(0, 100) for x in range (0,100)]
    list_divisible_three = [a for a in random_number_list2 if a % 3 == 0]
    
    length_of_random_list = len(random_number_list2)
    length_div_three = len(list_divisible_three)
    difference = length_of_random_list - length_div_three
    
    difference_list.append(difference)
difference_list
```

```python
#calculate avg_difference 
avg_difference = sum(difference_list)/len(difference_list)
avg_difference
```

```python
#get rid of new lines
multiline_text = multiline_text.replace('\n', "")
multiline_text
```

```python
#clean special characters and punctuation
cleaned_multiline = ""
for char in multiline_text:
    if char == " ":
        cleaned_multiline += char
    elif char.isalnum():
        cleaned_multiline += char
    else:
        cleaned_multiline += " "

cleaned_multiline
```

```python
#list of words from cleaned text
list_words = cleaned_multiline.split()
list_words
```

```python
#number of words in text
len(list_words)
```

```python
# number of unique words
unique_words_as_dict = dict.fromkeys(list_words)
len(list(unique_words_as_dict.keys()))
```

```python
# number of times unique words appeared
for word in list_words:
    if unique_words_as_dict[word] is None:
        unique_words_as_dict[word] = 1
    else:
        unique_words_as_dict[word] += 1

unique_words_as_dict
```

```python
#import zip_longest
from itertools import zip_longest
```

```python
#create function  to add header and line to dictionary
def return_dict_from_csv_line(header, line):
    zipped_line = zip_longest(header, line, fillvalue=None)
    ret_dict = {kv[0]: kv[1] for kv in zipped_line}
    return ret_dict
```

```python
# open file and create dictionary
with open('sales_record_LQ.csv', 'r') as fd:
    first_line = fd.readline()
    header = first_line.replace("\n", "").split(",")

    for i, line in enumerate(fd):
        line = line.replace("\n", "").split(",")
        d = return_dict_from_csv_line(header, line)
        print(d)
        if i > 10:
            break
```
