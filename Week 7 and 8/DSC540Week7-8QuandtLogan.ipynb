### Combine and Merge Datasets (you will have to either create a new dataset from your existing data or create a relationship between the data I have provided)

```python
#split dataframe into two, first one using first 6 columns
candy_df_merge_one = candy_df.iloc[:,:6]
candy_df_merge_one.head()
```

```python
#create 2nd data frame for merging, using remaining columns
candy_df_merge_two = candy_df.drop(['Q1: GOING OUT?', 'Q2: GENDER', 'Q3: AGE', 'Q4: COUNTRY', 'Q5: STATE, PROVINCE, COUNTY, ETC'], axis=1)
candy_df_merge_two.head()
```

```python
#merge dataframes back together using pd.merge on 'Internal ID'
candy_df_merged = pd.merge(candy_df_merge_one, candy_df_merge_two, on='Internal ID')
candy_df_merged.head()
```

```python
#read in 2016 file with timestamps
candy_df_2016 = pd.read_excel('candyhierarchy2016.xlsx')
candy_df_2016.head()
```

### Convert between string and date time

```python
#convert Timestamp column to str using strftime into new column
candy_df_2016['Timestamp_string'] = candy_df_2016['Timestamp'].apply(lambda x: x.strftime('%m/%d/%y'))
candy_df_2016.head()
```

```python
#check dtypes to confirm new column was converted as string
candy_df_2016.dtypes
```

### Generate Date Range

```python
#generate date range based off of timestamp using first and last occurences
candy_time_range = pd.date_range(candy_df_2016['Timestamp'].iloc[0], candy_df_2016['Timestamp'].iloc[-1], freq='H')
candy_time_range
```
